# 불필요한 컬럼 삭제


```python
df.drop(columns = ['환자번호', '국적', '환자정보', '조치사항', '이동경로', '등록일', '수정일', '노출여부'], inplace=True)
df
```

```python
df.dtypes
```

```python
# 확진일 -> datetime
df['확진일'] = pd.to_datetime(df['확진일'])
```

```python
# 지역의 공백제거
df['지역'].nunique()
```

```python
df['지역'].unique()
```

```python
df['지역'] = df['지역'].str.strip()
```

```python
# 지역, 타입을 category로 변경
df['지역'] = df['지역'].astype('category')
```

```python
df.info()
```

# 구별 확진자 동향
확진일 - 구별 확진자수 집계
pivotable 만들기

```python
df
```

```python
df_gu = pd.pivot_table(df, # 피벗할 테이블
                       index='확진일', # 행 위치에 들어갈 열
                       columns='지역', # 열 위치에 들어갈 열
                       values='연번', # 데이터로 사용할 열
                       aggfunc='count', # 데이터 집계함수
                       margins=True) 
df_gu
```

# 서울시 일별 추가 확진자 동향

```python
s_date = df_gu['All'][:-1]
s_date
```

```python
# 일별 추가 확진자가 많았던 순
s_date.sort_values(ascending=False)
```

```python
# 서울시 일별 추가 확진자 시각화
x = s_date.index
y = s_date.values
plt.plot(x, y)
plt.title('서울시 일별 추가 확진자 시각화')
plt.xlabel('확진일')
plt.ylabel('확진자 수')
plt.xticks(rotation=45)
plt.show()
```

# 서울시 구별 누적확진자 비교

```python
# 서울시 구별 누적 확진자가 많은 순으로
s_gu = df_gu.loc['All'][:-1]
s_gu = s_gu.sort_values(ascending=False)
s_gu
```

```python
# 서울시 구별 누적 확진자 시각화
x = s_gu.index
y = s_gu.values
plt.figure(figsize=(10,8))
plt.barh(x, y)
plt.title('서울시 구별 누적 확진자 시각화', size=10)
plt.xlabel('확진자 수')
plt.ylabel('자치구')
plt.show()
```
