# 2. Alice in Wonderland

```python
with open('./Alice.txt') as file:
    text = file.read()
```

```python
# said를 stopwords에 추가
sw_wc.add('said')
```

```python
wc = WordCloud(background_color='white',
              max_words=2000, stopwords=sw_wc)
wc = wc.generate(text)
```

```python
count = 0
for key, val in wc.words_.items():
    print(key, ':' , val)
    count += 1
    if count > 10:
        break
```

```python
plt.figure(figsize=(10,6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# 그림 위에 표시하기
mask = np.array(Image.open('Alice_mask.png'))
plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(background_color='white', mask=mask, 
              max_words=1000, stopwords=sw_wc).generate(text)
```

```python
plt.figure(figsize=(10,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

# 3. StarWars

```python
import random
```

```python
# HAN -> Han
# LUKE'S -> Luke
```

```python
# stop words
# int
# ext
```

```python
with open('./A_new_hope.txt') as file:
    text = file.read()
```

```python
text = text.replace('HAN', 'Han')
text = text.replace("LUKE'S", "Luke")
```

```python
sw = STOPWORDS
sw.add('int')
sw.add('ext')
```

```python
mask = np.array(Image.open('Stormtrooper_mask.png'))
plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(mask=mask, max_words=1000, stopwords=sw, margin=10,
              random_state=1).generate(text)
default_color = wc.to_array
```

```python
def grey_color_func(word, font_size, position, orientation, random_state=None,
                    **kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(60, 100)
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc.recolor(color_func=grey_color_func, random_state=2023),
           interpolation="bilinear")
plt.axis('off')
plt.show()
```
